Exploration of data generated by generate_data.py

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('synthetic_credit_card_transactions.csv')
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Plot the total amount spent per category
df.groupby('category')['amount'].sum().plot(kind='bar', ax=axes[0])
axes[0].set_title('Total Amount Spent per Category')
axes[0].set_ylabel('Amount')
axes[0].set_xlabel('Category')

# Plot the percentage of total spend per category
total_spend = df['amount'].sum()
grouped = df.groupby('category')['amount'].sum()
(grouped / total_spend).round(2).plot(kind='bar', ax=axes[1])
axes[1].set_title('Percentage of Total Spend per Category')
axes[1].set_ylabel('Percentage')
axes[1].set_xlabel('Category')

plt.tight_layout()
plt.show()
```
